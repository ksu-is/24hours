```python
import tkinter as tk
from tkinter import messagebox

# Create main window
root = tk.Tk()
root.title("24Hrs +")
root.geometry("300x450")

# Heading
heading_label = tk.Label(root, text="Welcome to 24Hrs +!", font=("Helvetica", 18))
heading_label.pack(pady=10)

# Frame for sign-up screen
frame_signup = tk.Frame(root)
frame_signup.pack(pady=20)

# Input fields
tk.Label(frame_signup, text="Full Name:").pack()
entry_name = tk.Entry(frame_signup)
entry_name.pack()

tk.Label(frame_signup, text="Phone Number:").pack()
entry_phone = tk.Entry(frame_signup)
entry_phone.pack()

tk.Label(frame_signup, text="Email Address:").pack()
entry_email = tk.Entry(frame_signup)
entry_email.pack()

tk.Label(frame_signup, text="Password:").pack()
entry_signup_password = tk.Entry(frame_signup, show="*")
entry_signup_password.pack()

tk.Label(frame_signup, text="Confirm Password:").pack()
entry_confirm_password = tk.Entry(frame_signup, show="*")
entry_confirm_password.pack()

tk.Label(frame_signup, text="School/Occupation:").pack()
entry_school_occupation = tk.Entry(frame_signup)
entry_school_occupation.pack()

from tkinter import Menu, IntVar, Menubutton  # Make sure this is at the top with other imports

# Function to open home page
def open_home_page(name):
    home = tk.Toplevel(root)
    home.title("Home Page")
    home.geometry("300x400")

    label = tk.Label(home, text=f"Welcome, {name}!", font=("Helvetica", 18))
    label.pack(pady=20)

    # Exit button
    button = tk.Button(home, text="Exit", command=home.destroy)
    button.pack(pady=10)

    # Menu button setup
    menubutton = Menubutton(home, text="â‰¡", relief="raised", direction="below")
    menubutton.menu = Menu(menubutton, tearoff=0)
    menubutton["menu"] = menubutton.menu

    var1 = IntVar()
    var2 = IntVar()
    var3 = IntVar()

    menubutton.menu.add_checkbutton(label="Home", variable=var1)
    menubutton.menu.add_checkbutton(label="Stats", variable=var2)
    menubutton.menu.add_checkbutton(label="Bio", variable=var3)
    
    menubutton.pack(pady=10)


# Function for sign-up validation
def signup():
    name = entry_name.get()
    phone_number = entry_phone.get()
    email_address = entry_email.get()
    password = entry_signup_password.get()
    confirm_password = entry_confirm_password.get()
    school_occupation = entry_school_occupation.get()
    
    if password != confirm_password:
        messagebox.showerror("Error", "Passwords do not match!")
        return
    
    if len(password) < 5 or not any(c.isupper() for c in password):
        messagebox.showerror("Error", "Password must be at least 5 characters and contain at least one uppercase letter.")
        return
    
    if name and phone_number and email_address and password and school_occupation:
        print("Sign Up with Name:", name, "Phone:", phone_number, "Email:", email_address, "School/occupation:", school_occupation)
        messagebox.showinfo("Sign Up", "Sign Up Successful!")
        open_home_page(name)
    else:
        messagebox.showerror("Error", "Please fill in all fields!")

# Sign-up button
signup_button = tk.Button(frame_signup, text="Create Account", command=signup, width=15)
signup_button.pack(pady=15)

# Run the app
root.mainloop()

```
